```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data", header=None)
df.describe()
```

```python
# Menghapus baris yang memiliki data tak lengkap
df = df[df[0:].notna()]
df
```

```python
# Merubah nama atribut kolom
df = df.rename(columns={0: "Age",
          1: "Sex",
          2: "Height",
          3: "Weight",
          4: "QRS duration",
          5: "P-R interval",
          6: "Q-T interval",
          7: "T interval",
          8: "P interval",
          14: "Heartrate",
          279: "Class"})
df
```

```python
df.columns
```

```python
# Drop kolom yang tidak dipakai
df = df.drop(pd.concat([df.iloc[:,9:14],df.iloc[:,15:279]]),axis=1)
df
```

```python
# Menghilangkan data yang memiliki value ?
df = df[df["Heartrate"] != "?"]
df
```

```python
# mengacak urutan
df = df.sample(451).reset_index(drop=True)
```

```python
# Merubah tipe data kolom heartrate dari object menjadi integer
df["Heartrate"] = pd.to_numeric(df["Heartrate"])
if "index" in df.columns:
    df = df.drop("index", axis=1)

# Membuat 3 fold
fold1 = (df.iloc[0:150].reset_index(drop=True), df.iloc[150:450].reset_index(drop=True))
fold2 = (df.iloc[150:300].reset_index(drop=True), pd.concat([df.iloc[300:450], df.iloc[0:150]]).reset_index(drop=True))
fold3 = (df.iloc[300:450].reset_index(drop=True), df.iloc[0:300].reset_index(drop=True))

# Cek isi fold 2
test, train = fold2
print(train)
```

```python
# Fungsi normalisasi (merubah angka menjadi range antara 1-0)
def norm(df):
  df = (df - df.min()) / (df.max() - df.min())
  return df

# Cek apakah fungsi bekerja dengan baik
X = df.drop("Class", axis=1)
Y = df.Class

X = norm(X)
print(X)
```

```python
# Fungsi menghitung jarak euclidean antara 2 titik
def euclidean(x1, x2):
  return np.sqrt(np.sum((x1-x2)**2))

euclidean(X.iloc[0],X.iloc[1])
```

```python
# Fungsi kNN
def knn(X_train, Y_train, X_test, k):
  # Membandingnkan kedekatan nilai test dengan data training
  dist=[]
  for row in range(X_train.shape[0]):
    dist.append(euclidean(X_train.iloc[row], X_test))

  data = X_train.copy()
  data['Dist'] = dist
  data['Class'] = Y_train # Membuat dataframe berisi hasil euclidean dengan value class
  data = data.sort_values(by='Dist').reset_index(drop=True) # Mengurutkan dari nilai euclidean paling besar (paling dekat)

  Y_pred = data.iloc[:k].Class.mode() # Ambil k kelas paling atas
  return Y_pred[0]

```

```python
# Fungsi cek akurasi, output = persentasi kebenaran pemberian class
def acc(Y_pred, Y_true):
  true = 0
  for i in range(len(Y_pred)):
    if Y_pred[i] == Y_true[i]:
      true+=1
  return true/len(Y_pred)
```

```python
# Fungsi utama
def evaluate(fold, k):
    test, train = fold # Memisahkan antara train data dan test data
    X_train, Y_train = train.drop('Class', axis=1), train.Class # Memisahkan tiap data menjadi class (result) dan data perbandingan
    X_test, Y_test = test.drop('Class', axis=1), test.Class
    X_train = norm(X_train) # normalisasi data perbandingan
    X_test = norm(X_test)
    Y_preds = [] # list untuk menyimpan hasil output
    for row in range(X_test.shape[0]):
      Y_preds.append(knn(X_train, Y_train, X_test.iloc[row], k)) # menyimpan hasil output
    
    return (acc(Y_preds, Y_test)) # mengembalikan akurasi
```

```python
# Tes running untuk k = 5
k = 5
accs = []
folds = [fold1, fold2, fold3]
for i in range(len(folds)):
  accs.append(evaluate(folds[i], k))
print(f'Menggunakan k : {k}, dengan rata-rata akurasi : {sum(accs)/3}')
```

```python
# Tes running untuk k dari 1-10
listOfAcc = []
for k in range(1,11):
    accs = []
    for i in range(len(folds)):
        accs.append(evaluate(folds[i],(k)))
    print(f'Menggunakan k : {k}, dengan rata-rata akurasi : {sum(accs)/3}')
    listOfAcc.append(sum(accs)/3)
```

```python
# Persiapan data untuk plot
dfAcc = pd.DataFrame(columns=['Accuracy'], data=listOfAcc)
dfAcc.index += 1
```
