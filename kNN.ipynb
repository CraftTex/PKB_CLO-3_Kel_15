```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data", header=None)
df.describe()
```

```python
# Menghapus baris yang memiliki data tak lengkap
df = df[df[0:].notna()]
df
```

```python
# Merubah nama atribut kolom
df = df.rename(columns={0: "Age",
          1: "Sex",
          2: "Height",
          3: "Weight",
          4: "QRS duration",
          5: "P-R interval",
          6: "Q-T interval",
          7: "T interval",
          8: "P interval",
          14: "Heartrate",
          279: "Class"})
df
```

```python
def norm(df):
  df = (df - df.min()) / (df.max() - df.min())
  return df

df["Heartrate"] = pd.to_numeric(df["Heartrate"])
X = df.drop("Class", axis=1)
Y = df.Class

X = norm(X)
print(X)
```
